```python
# 실습과제 -1 : 네이버 뉴스 사이트에 검색어로 '코로나'를 입력하여  출력되는
# 뉴스 내용을 5개 출력해주는 코드를 구현해보세요

import requests
from bs4 import BeautifulSoup

base_url = "https://search.naver.com"

# 네이버 뉴스 사이트 : 검색어로 '코로나'를 입력하여 아래 주소를 얻어온다
# https://search.naver.com/search.naver?query=%EC%BD%94%EB%A1%9C%EB%82%98&where=news&ie=utf8&sm=nws_hty

search_url = base_url + "/search.naver?query=%EC%BD%94%EB%A1%9C%EB%82%98&where=news&ie=utf8&sm=nws_hty"

resp = requests.get(search_url)
html_src = resp.text
soup = BeautifulSoup(html_src, 'html.parser')


news_items = soup.select('ul[class="type01"] > li')
print(len(news_items))  # 10
# print(news_items[0])
print('\n')

# 크롬 개발자 도구를 사용한다
for item in news_items[:5]:
    link = item.find('a', attrs={'class':'_sp_each_title'}).get('href')
    news_link = link[:]
    print(news_link) # url link 를 찾음
   
    news_title = item.find('a', attrs={'class':'_sp_each_title'}).getText()
    print(news_title)

    news_content = item.find_all('dd')
    for contents in news_content:
        print(contents.text)
    print('\n')
```

```python
# 실습과제 - 2  : 다음 뉴스 사이트에 검색어로 '코로나'를 입력하여  출력되는
# 뉴스 내용을 5개 출력해주는 코드를 구현해보세요

base_url = "https://search.daum.net"

# 다음 뉴스 사이트 : 검색어로 '코로나'를 입력하여 아래 주소를 얻어온다
# https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%BD%94%EB%A1%9C%EB%82%98
search_url = base_url + "/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EC%BD%94%EB%A1%9C%EB%82%98"

resp = requests.get(search_url)
html_src = resp.text
soup = BeautifulSoup(html_src, 'html.parser')


news_items = soup.select('div[class="wrap_cont"]')
print(len(news_items))  # 10
# print(news_items[0])
print('\n')

# 크롬 개발자 도구를 사용한다
for item in news_items[:5]:
    link = item.find('a', attrs={'class':'f_link_b'}).get('href')
    news_link = link[:]
    print(news_link) # url link 를 찾음
   
    news_title = item.find('a', attrs={'class':'f_link_b'}).getText()
    print(news_title)

    news_content = item.find_all('span',attrs={'class':'f_nb date'})
    for contents in news_content:
        print(contents.text)
        
    news_content = item.find_all('p',attrs={'class':'f_eb desc'})
    for contents in news_content:
        print(contents.text)       
    print('\n')
```
